## 1. Load Libraries & Data

Import any necessary libraries and load the remote CSV file below into a DataFrame. **Resolve any import issues you need to.**

```python
import pandas as pd


url = "https://s3.amazonaws.com/ds1002-resources/data/ftx-2021-01-01--2023-12-31.csv"


df = pd.read_csv(url, sep=';')

df.head(10)
```

## 2. Get Summary Data

In code, display how many rows and columns are in the
raw dataset.

```python
num_rows = df.shape[0]
num_columns = df.shape[1]

print(num_rows)
print(num_columns)
# 741 rows & 11 columns
```

## 3. Clean and Organize the Data

Check for data quality.

- Resolve any empty values or duplicate rows.
- Order the data properly if necessary.
- Drop the following columns and save the dataframe: `timeOpen`, `timeClose`, `timeHigh`, `timeLow`, `marketCap`



```python
duplicate_rows = df[df.duplicated()]
df.drop_duplicates(inplace=True)
droppings = ['timeOpen', 'timeClose', 'timeHigh', 'timeLow', 'marketCap']
df.drop(droppings, axis=1, inplace=True)
```

Now display the first 10 rows of the cleaned dataset.

```python
df.head(10)
```

## 4. Calculate Daily Gains/Losses

Using Pandas, add a new column named `["close_change"]` to the dataframe that calculates the difference between daily closing prices for FTX. This will result in positive (growth) and negative (loss) values.

Note:
- Be careful about how you calculate this. Each row should calculate that difference from the row above it, which must be the previous calendar day.
- Hint: This is not complicated and does not require a for-loop.  See the Pandas review notebook for reference.

```python
df["close_change"] = df['close'].pct_change()
```

## 5. Identify Growth Days

Based on the daily change in closing prices calculated above, identify the top 10 days of positive growth for FTX.

Display your results in a table showing both the rate of change and their dates.

```python
df['day_change'] = df['close'].diff()

positive_growth = df[df['day_change'] > 0]
#not sure why I need the df nested in the other df but the colab helper recommended this to me and that made it work
top10 = positive_growth.sort_values(by='day_change', ascending=False)
#Ascending=False was a later addition, makes more sense for the data to be presented in that manner
top10.head(10)
```

## 6. Calculate the Percentage Rate of Change

Next, create a new column named `["rate_of_change"]` based on `["close_change"]`.

Use the Pandas Review notebook for reference. Again, this is an easy operation in Pandas and does not require a for-loop.

```python
df['rate_of_change'] = df['close_change'].pct_change()
df.head(10)
```

## 7. Identify Days of Greatest Percentage Change

Based on the rate change column calculated above, identify the top 5 days of greatest positive rate change, and the top 5 days of greatest negative rate change.

Display your results in a table showing both the rate of change and their dates.

```python
#top 5 positive change
positive_change = df[df['rate_of_change'] > 0]
top_5_positive_change = positive_change.sort_values(by='rate_of_change', ascending=False)
#top 5 negative change
negative_change = df[df['rate_of_change'] < 0]
top_5_negative_change = negative_change.sort_values(by='rate_of_change', ascending=True)
#presenting both
print(top_5_positive_change.head(5))
print(top_5_negative_change.head(5))
```

## 8. Plot the Data

Create two separate plots of the data -- one for the daily closing price and one for the daily gains/losses. These can be histograms or line plots.

```python
import matplotlib.pyplot as plt
import numpy as np
# x-axis values was giving me the full UTS readout for every day making it incoherent, decided to use numpy insetad
tick_locations = np.arange(0, 741, 50)
tick_labels = range(0, 741, 50)
plt.figure(figsize=(10, 6))
plt.plot(df['timestamp'], df['close'], color='blue', label='Daily Closing Price')
plt.title('Change in Daily Closing Price')
plt.xlabel('date')
plt.ylabel('Closing Price in USD')
plt.xticks(tick_locations, tick_labels, rotation=45)
plt.legend()
plt.show()
```

```python
# had to do the same thing for x-axis here too
tick_locations = np.arange(0, 741, 50)
tick_labels = range(0, 741, 50)
plt.figure(figsize=(10, 6))
plt.plot(df['timestamp'], df['day_change'], color='green', label='Daily Gain/Loss')
plt.title('Daily Gain/Loss')
plt.xlabel('Date')
plt.ylabel('Gain/Loss in USD')
plt.xticks(tick_locations, tick_labels, rotation=45)
plt.legend()
plt.show()
```

Combine the two plots from above into a single plot, displaying both together. Your plot should have two Y-axis scales, should differentiate the two values by color, and should include a legend.

```python
tick_locations = np.arange(0, 741, 50)
tick_labels = range(0, 741, 50)
# close change
fig, ax1 = plt.subplots(figsize=(10, 6))
ax1.plot(df['timestamp'], df['close'], color='blue', label='Daily Closing Price')
ax1.set_xlabel('Date')
ax1.set_ylabel('Closing Price ($)', color='blue')

# daily change
ax2 = ax1.twinx()
ax2.plot(df['timestamp'], df['day_change'], color='green', label='Daily Gain/Loss')
ax2.set_ylabel('Gain/Loss in USD', color='green')

# make x-axis legible
plt.xticks(tick_locations, tick_labels, rotation=45)

plt.title('Daily Closing Price and Gain/Loss')
plt.show()
```
